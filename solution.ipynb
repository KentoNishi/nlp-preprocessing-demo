```python
# import libraries we need
import re  # regular expressions for splitting strings
import nltk  # natural language toolkit
from nltk import word_tokenize  # tokenize sentences
from nltk.corpus import stopwords  # list of stopwords
from nltk.stem.porter import PorterStemmer  # stemming algorithm
```

```python
# reading the file
with open("text.txt", "r") as file:  # read text.txt
    text = file.read()
print(text)
```

```python
# remove punctuation marks and such
tokens = word_tokenize(text)  # split the string into a list of words
words = [word for word in tokens if word.isalpha()]  # only select tokens with letters
print(words)
```

```python
# remove stopwords (nonessential words)
nltk.download("punkt")  # download stopwords
stop_words = set(stopwords.words("english"))  # read stopwords
print(stop_words)
words = [word for word in words if not word in stop_words]  # only select non-stopwords
print(words)
```

```python
# apply stemming to convert them into their base forms
porter = PorterStemmer()  # initialize a stemmer instance
words = [porter.stem(word) for word in words]  # stem all words
print(words)
```
